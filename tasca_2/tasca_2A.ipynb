#### Exercici 4
Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : "les matrius es poden transmetre / broadcast

entre si les seves dimensions coincideixen o si una de les matrius té una mida d'1".

```python
# Operacións: 

# Matriu (4x5) i matriu (1x5), sumem

a = np.random.randint(10, size=(4,5))
print(f'a: {a}')
b = np.full(shape=(1,5), fill_value=3)
print(f'b: {b}')
print(f'A+B -> broadcast files de b')
print(a+b)
```

```python
# fem b una matriu de (4,) i multipliquem

b = np.full(shape=(4,), fill_value=2)
print(f'b: {b}')
print(f'a+b -> broadcast de b')
try:
    print(a+b)
except ValueError:
    print('Genera un error de ValueError perquè els shapes no coincideixen')
    print('Canviem el shape de b a (4,1) i tornem a sumar')
    print(a+b.reshape(-1,1))
    print(f'shapes: a: {a.shape} , b: {b.shape}')

print(a*b.reshape(-1,1))
```

```python
# Fem una matriu de 5x1x7x3 i una matriu de 4x1x3 i veiem el resultat
a = np.random.randint(10, size=(2,1,3,3))
b = np.random.randint(10, size=(2,1,3))
print(f'a: {a}')
print(f'b: {b}')
print(f'a+b: shape: {(a+b).shape} \n{a+b}')
```

```python
# Broadcasting senzill
a = np.random.randint(5, size=(5,3))
b = 3
print(a+b)
print(a*b)
```

```python
b = np.array([[3,4,3]])
print(f'a: {a.shape} \nb: {b.shape}')
print(a*b)
```
